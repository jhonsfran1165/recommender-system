```python
import pandas as pd


def read_in_chunks(file_object, chunk_size=1024, sep='*', header=None):
    csv_reader = pd.read_csv(file_object, sep=sep, iterator=True, chunksize=chunk_size, header=header)
    first_chunk = csv_reader.get_chunk()
    chunk = pd.DataFrame(first_chunk)
    for l in csv_reader:
        id = l.iloc[0,0]
        yield chunk
        chunk = pd.DataFrame(l)

chunk_iter = read_in_chunks("modified.csv", header=0)

for chunk in chunk_iter:
    print("_________")
    print(chunk.iloc[:,:1])
    
# df = pd.read_csv(
  #      'modified.csv',
  #  encoding = "ISO-8859-1",
  #  sep='*',
  #  engine='python',
  #  doublequote=True
#)

#df
```

```python
import pandas as pd

df = pd.read_csv(
        '../data/copies.txt',
    encoding = "ISO-8859-1",
    sep='|',
    engine='python',
    doublequote=True
)

df.columns = df.columns.str.replace('|','')
df.columns = df.columns.str.replace("''",'|')
df.columns = df.columns.str.replace("'","")
columns = df.columns.str.split("|")[0]
df.columns = [("data")]

new_df = df['data'].str.split(pat="|", expand=True)
df[columns] = new_df.iloc[:,:9]
df = df.iloc[:,1:10]
```

```python
df.set_index('C.COPYNO', inplace=True)
df.sort_values('C.COPYNO')
df
```
