```python
import pandas as pd

df = pd.read_csv(
        '../data/copies.txt',
    encoding = "ISO-8859-1",
    sep='|',
    engine='python',
    doublequote=True
)

df.columns = df.columns.str.replace('|','')
df.columns = df.columns.str.replace("''",'|')
df.columns = df.columns.str.replace("'","")
columns = df.columns.str.split("|")[0]
df.columns = [("data")]

new_df = df['data'].str.split(pat="|", expand=True)
df[columns] = new_df.iloc[:,:9]
df = df.iloc[:,1:10]
```

```python
df.set_index('C.COPYNO', inplace=True)
df.sort_values('C.COPYNO')
df
```
