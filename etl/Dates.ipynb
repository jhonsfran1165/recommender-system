```python
# key(19990812), date(12/08/99), data string (1999-08-12), year, month, week, day, semester, quarter, dayofmonth, dayofweek, dayof year

# https://pandas.pydata.org/docs/reference/api/pandas.Series.dt.dayofyear.html

import pandas as pd

def create_date_table(start='1990-01-01', end='2020-12-31'):
    df = pd.DataFrame({"Date": pd.date_range(start, end)})
    df["key"] = df.Date.dt.strftime('%Y%m%d').astype(int)
    df["Date_format"] = df.Date.dt.strftime('%m/%d/%Y').astype(str)
    df["Day_Week"] = df.Date.dt.dayofweek.astype(int)
    df["Day_Name"] = df.Date.dt.day_name().astype(str)
    df["Day_Month"] = df.Date.dt.day.astype(int)
    df["Day_Year"] = df.Date.dt.day_of_year.astype(int)
    df["Month"] = df.Date.dt.month.astype(int)
    df["Month_Name"] = df.Date.dt.month_name().astype(str)
    df["Days_in_Month"] = df.Date.dt.daysinmonth.astype(int)
    df["Week"] = df.Date.dt.isocalendar().week.astype(int)
    df["Quarter"] = df.Date.dt.quarter.astype(int)
    df["Year"] = df.Date.dt.year.astype(int)
    df["Semeter"] = (df.Quarter + 1) // 2
    return df

df = create_date_table()
```

```python
df.set_index('key', inplace=True)
df.sort_values('key')
df
```
